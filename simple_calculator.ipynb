```python
#simple calculator
num = int(input("enter amount:"))
smd = input("Multiplication, division, addition, subtraction(*,/,+,-):")
num1 = int(input("enter amount:"))

if smd == "+":
    print(num+num1)
elif smd == "-":
    print (num-num1)
elif smd == "*":
    print (num*num1)
```

```python
#create a countdown timer
import time
def countdown(t):
    while t > 0:
        print(t)
        t -= 1
        time.sleep(1)
    print("Stop!")
countdown (int(input("enter second:")))
```

```python
#create a countdown timer
import time
def countdown(time_sec):
    while time_sec:
        min, sec = divmod (time_sec, 60)
        timeformat = '{:02d}:{:02d}'.format(min,sec)
        print(timeformat, end='\r')
        time.sleep(1)
        time_sec -= 1
    print("Stop!")
countdown (int(input("enter second:")))
```
