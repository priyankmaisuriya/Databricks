```
#Count Alphabets, Digits and Special Characters in a String

email="priyank!@ 12 cd 1212"
alpha=digit=special=0

for i in range(len(email)):
    if((email[i] >= 'a' and email[i] <= 'z') or (email[i] >= 'A' and email[i] <= 'Z') ):
        alpha = alpha + 1
    elif(email[i] >= '0' and email[i] <= '9'):
        digit = digit + 1
    else:
        special = special + 1
print("Alphabets:",alpha)
print("Digits:",digit)
print("Special Characters:",special)
```

```
#Replace Blank Space with Hyphen in a String

intro1 = "priyank maisuriya from rampura"
intro=''

for i in range(len(intro1)):
    if(intro1[i]==' '):
        intro=intro + '_'
    else:
        intro = intro + intro1[i] 
print("Update:",intro)
```

```
n = 371
s = 0
t = n

while t > 0:
    digit = t % 10
    s += digit ** 3
    t //= 10
if n == s:
    print(n,"is an Armstrong number")
else:
    print(n,"is not an Armstrong number")
```

```
#Print the Pascal’s triangle for 4 number of rows

rows = 4
for i in range(0, rows):
    trig = 1
    for j in range(1, rows-i):
        print("  ", end="")
    
    for x in range(0,i+1):
        print("  ", trig, end="")
        trig = int(trig * (i - x) / (x + 1))
    print()
```
