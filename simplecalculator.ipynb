```python
#simple calculator
def add(num,num1):
    return num + num1
def subtract(num, num1):
    return num - num1
def multiply(num, num1):
    return num * num1
def divide(num, num1):
    return num / num1
print()
smd = input("Multiplication, division, addition, subtraction(*,/,+,-):")

num = int(input("enter amount first:"))
num1 = int(input("enter amount second:"))

if smd == "+":
    print(add(num,num1))
elif smd == "-":
    print (subtract(num, num1))
elif smd == "*":
    print (multiply(num, num1))
elif smd == "/":
    print (divide(num, num1))
```
