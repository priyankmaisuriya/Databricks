```
#calculate sum of even number from 1 to 50 using for loop
max=50
total=0

for num in range(1,max+1):
    if(num % 2==0):
        print("{0}".format(num))
        total=total+num
print("sum of even number from 1 to {0} = {1}".format(num,total))
```

```
#check leap year or not
year=2020
if (year % 4) == 0:
    print(year,"is a leap year")
elif (year % 100) ==0:
    print(year,"is not a leap year")
elif (year % 400) ==0:
    print(year,"is leap year")
else:
    print(year,"is not a leap year")
```

```
#return largest number in a list
list1 = [70,80,120,87,46]
for i in range(len(list1)):
     for j in range(i+1,len(list1)):
            if list1[i]>list1[j]:
                 list1[i],list1[j]=list1[j],list1[i]
print("Largest number", list1[-1])         
```
