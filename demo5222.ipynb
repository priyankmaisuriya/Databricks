```python
#palindrome or not
a="malayalam"
b=a[-1::-1]

if (a == b):
    print("yes")
else:
    print("no")
```

```python
rows=5
for i in range(1, rows + 1):
    for j in range(1, rows + 1):
        if(j < i):
            print(' ', end = '  ')
        else:
            print('*', end = '  ')
    print()
```
